```python
import pandas as pd

# Create a DataFrame
data = {'Category': ['A', 'A', 'B', 'B', 'A', 'B'],
        'Value': [1, 2, 3, 4, 5, 6],
        'Type': [0, 1, 1, 1, 0, 0]}
df = pd.DataFrame(data)

# Group the DataFrame by 'Category'
grouped = df.groupby('Category')

# Define a custom function to calculate the mean value for each group
def imputed_forms_custom_mean(group):
    # print(group)
    return pd.Series({
        'Value': group.loc[group['Type'] == 0, 'Value'].mean().astype(int),
        'Type': 0})
    # return value

# Apply the custom function to calculate the mean for each group
result = grouped.apply(imputed_forms_custom_mean).reset_index()

print(result)
```
